```python
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from scipy import stats

np.random.seed(1)
X = np.random.normal(0,1,100)
eps = np.random.normal(0, 0.25, 100)

Y = [( -1 + X[i]*0.5 + eps[i]) for i in range(100)]

data = pd.DataFrame({'x': X, 'y': Y})

linear_model = smf.ols('y ~ x', data).fit()
print(linear_model.summary())
intercept = linear_model.params.Intercept
slope = linear_model.params.x

Y_reg = [intercept + slope * X[i] for i in range(100)]

X_p = np.linspace(min(X), max(X), 100)
Y_p = [ -1 + 0.5 * X_p[i] for i in range(100)]


plt.scatter(X, Y)
plt.plot(X, Y_reg, c = 'red', label = "Least Squares Line")
plt.plot(X_p, Y_p, c = 'green', label = "Population")

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
poly_model = smf.ols('y ~ x + I(x**2)', data).fit()
print(poly_model.summary())
poly_model.params
```

```python
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)

data = pd.DataFrame({'x1': x1, 'x2' : x2, 'y': y})

linear_model = smf.ols('y ~ x1 + x2', data).fit()
print(linear_model.summary())

print(np.corrcoef(x1, x2))

print("B1 ", linear_model.params.x1)
print("B2 ", linear_model.params.x2)

plt.scatter(x1, x2)
plt.show()
```

```python
data = pd.DataFrame({'x1': x1, 'y': y})

linear_model = smf.ols('y ~ x1', data).fit()
print(linear_model.summary())

print("B1 ", linear_model.params.x1)

plt.scatter(x1, y)
plt.show()
```

```python
data = pd.DataFrame({'x2': x2, 'y': y})

linear_model = smf.ols('y ~ x2', data).fit()
print(linear_model.summary())

print("B2 ", linear_model.params.x2)

plt.scatter(x2, y)
plt.show()
```

# Zad 11

```python
df = pd.read_csv("Auto.csv", na_values="?")
df = df.dropna()

df["origin"] = df["origin"].astype("category")

df.isnull().sum()
```

```python
len_train = int(0.8 * len(df))
train = df[:len_train]
validation = df[len_train:]

model = smf.ols("mpg ~ horsepower", data = train).fit()
print(model.summary())
```

```python

pd.plotting.scatter_matrix(df)
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
model = smf.ols("mpg ~ horsepower + weight + year + cylinders + origin", data=train).fit()
print(model.summary())

model2 = smf.ols("mpg ~ I(np.log(horsepower)) + weight + year + cylinders + origin", data=train).fit()
print(model2.summary())

model3 = smf.ols("mpg ~ horsepower * weight + year + cylinders + origin", data=train).fit()
print(model3.summary())
```

```python
from sklearn.metrics import mean_squared_error

preds = model.predict(validation)
mse = mean_squared_error(validation["mpg"], preds)

print(mse)

preds = model2.predict(validation)
mse = mean_squared_error(validation["mpg"], preds)

print(mse)

preds = model3.predict(validation)
mse = mean_squared_error(validation["mpg"], preds)

print(mse)
```
